
# **Arithmetic Series**
Modified from Arithmetic Series - [Wolfram MathWorld](https://mathworld.wolfram.com/ArithmeticSeries.html)

An arithmetic sequence $\{a_k\}$ for $k=1, 2, ..., $ in which each term is computed from the previous one by adding (or subtracting) a constant $d$.

\begin{align*} a_1 & \\ a_2 &\ \ =\ \ a_1 + d \\ a_3 &\ \ =\ \ a_2 + d \ \ =\ \ a_1 + 2d \\ a_4 &\ \ =\ \ a_3 + d \ \ =\ \ a_2 + 2d \ \ =\ \ a_1 + 3d \\ \vdots\\ a_k &\ \ =\ \ a_{(k-1)}+d \ =\ a_{(k-2)}+2d \ =\ \ldots \ =\ \ a_1+(k-1)d \end{align*}
Therefore, for $k>=1$,

$$ a_k = a_1 + (k-1)d$$
With $a_1$ and $d$ given.



The sequence of the partial sum of the first $n$ terms of a sequence is called a series, in the case of an arithemetic sequence the series is called an arithmetic series.

If $a_1= 1$ and $d=1$ we have, the simplest arithmnetic sequence, the natural numbers sequence $\{a_k = k \}$ for $k=1, 2, ..., $ i.e. the seuqnece $$1, 2, ...,$$ The corresponmding series is $$ S_n = 1 + 2+ \ldots + (n-1)+ n = \sum_{k=1}^n k $$ can be calculated as a function of $n$.

Assume that we allign one $S_{n-1}$ with a second $S_{n-1}$ that have the terms adden reverse order and you add term to term, we have \begin{array}{llclclclclc} S_{n-1} &=& 1 &+& 2 &+& \ldots &+& (n-2) &+& (n-1) \\ S_{n-1} &=& (n-1) &+& (n-2) &+& \ldots &+& 2 &+& 1 \\ \hline \\ 2S_{n-1} &=& n &+& n &+& \ldots &+& n &+& n \\ \end{array} so $$ 2S_{n-1} = (n-1)n $$ then $$ S_{n-1} = \frac{1}{2} (n-1)n$$ therefore $$ S_{n} = \frac{1}{2} n(n+1)$$

\begin{align*} S_n &= \sum_{k=1}^n a_k \\ &= \sum_{k=1}^n[a_1+(k-1)d] \\ &= na_1 + d \sum_{k=1}^n(k-1) \\ &= na_1 + d \sum_{k=0}^{n-1} k \\ &= na_1 + d \sum_{k=1}^{n-1} k \\ &= na_1 + \frac{1}{2} d(n-1)n\\ \end{align*}
so $$ S_n = \frac{1}{2} n[2 a_1 +d(n-1)].$$

Note, however, that $$a_1+a_n=a_1+[a_1+d(n-1)]=2a_1+d(n-1),$$
so $$S_n = \frac{1}{2}n(a_1+a_n),$$

or $n$ times the arithmetic mean of the first and last terms! This is the trick Gauss used as a schoolboy to solve the problem of summing the integers from $1$ to $100$ given as busy-work by his teacher. While his classmates toiled away doing the addition longhand, Gauss wrote a single number, the correct answer

$$\frac{1}{2}(100)(1+100)=50Â·101=5050$$
 
on his slate (Burton 1989, pp. 80-81; Hoffman 1998, p. 207). When the answers were examined, Gauss's proved to be the only correct one.



If the sequence starts at index $0$, $\{a_k\}$ for $ k=0, 1, 2, ..., $ we have

\begin{align*} a_0 & \\ a_1 &\ \ =\ \ a_0 + d \\ a_2 &\ \ =\ \ a_1 + d \ \ =\ \ a_0 + 2d \\ a_3 &\ \ =\ \ a_2 + d \ \ =\ \ a_1 + 2d \ \ =\ \ a_0 + 3d \\ \vdots\\ a_k &\ \ =\ \ a_{(k-1)}+d \ =\ a_{(k-2)}+2d \ =\ \ldots \ = \ \ a_1+(k-1)d \ \ =\ \ a_0+kd \end{align*}

for given $a_0$ and $d$.

In this case the sum of the first $n$ terms of the arithemnetic sequence is then given by \begin{align*} S_n &= \sum_{k=0}^{n-1} a_k \\ &= \sum_{k=0}^{n-1}[a_0+kd] \\ &= n a_0 + d \sum_{k=0}^{n-1} k \\ &= n a_0 + d \sum_{k=1}^{n-1} k \\ &= n a_0+ \frac{1}{2} d(n-1)n \\ \end{align*} so $$ S_n = \frac{1}{2} n[2 a_0 + d(n-1)].$$

Geometric Series
Modified from Arithmetic Series - Wolfram MathWorld
An geometric series is the sum of a sequence $\{a_k\}$ for $k=1, 2, ..., $ in which each term is computed from the previous one by adding (or subtracting) a constant $d$.

\begin{align*} a_0 &\ \ =\ \ r^0 \ \ =\ \ 1\\ a_1 &\ \ =\ \ a_0 * r \ \ =\ \ r\\ a_2 &\ \ =\ \ a_1 * r\ \ =\ \ a_0 * r^2 \ \ =\ \ r^2 \\ a_3 &\ \ =\ \ a_2 * r \ \ =\ \ a_0 * r^3 \ \ =\ \ r^3 \\ a_4 &\ \ =\ \ a_3 * r\ \ =\ \ a_0 * r^4 \ \ =\ \ r^4 \\ \vdots\\ a_k &\ \ =\ \ a_{(k-1)}* r \ =\ a_0 * r^k \ \ =\ \ r^k \end{align*}
The sum of the first $n$ terms of the geometric sequence (called an geometric serie) is then given by $$ \begin{array}{ll} S_n &= 1 + r + r^2 + \cdots + r^{n-1} +r^n \\ &= (1 + r + r^2 + \cdots + r^{n-1} ) +r^n \\ &= S_{n-1} +r^n \end{array} $$

$$ \begin{array}{ll} S_n &= 1 + r + r^2 + \cdots + r^{n-1} +r^n \\ &= 1 + (r + r^2 + \cdots + r^{n-1} +r^n) \\ &= 1 + r (1 + r + r^2 + \cdots + r^{n-1} ) \\ &= 1 + r S_{n-1} \end{array} $$
Then $$ S_{n-1} +r^n = 1 + r S_{n-1} $$

Therefore $$ S_{n-1} (1 -r) +r^n = 1 - r^n $$

$$ S_{n-1} = \frac{1 - r^n }{1 -r} $$$$ S_{n} = \frac{1 - r^{n+1} }{1 -r} = \frac{ r^{n+1} - 1}{r -1} $$

Hay dos tipos de sumas de potencia que se consideran comÃºnmente. La primera es la suma de las p-Ã©simas potencias de un conjunto de n variables $x_k$,

$$S_{p}(x_{1}, ..., x_{n}) = \sum_{n}^{k=1}x_{k}^{p})$$
y el segundo es el caso especial $x_k=k$, i.e.,

$$S_{p}(n)= \sum_{n}^{k=1}k^{p})$$
Las sumas generales de potencia surgen comÃºnmente en las estadÃ­sticas. Por ejemplo, las estadÃ­sticas k se definen mÃ¡s comÃºnmente en tÃ©rminos de sumas de potencia. Las sumas de potencia estÃ¡n relacionadas con polinomios simÃ©tricos mediante las fÃ³rmulas de Newton-Girard.

La suma de k multiplicada por la k-Ã©sima potencia de x viene dada analÃ­ticamente por

$$\sum_{k=0}^{n}kx^{k} = \frac{x-(n+1)x^{n+1}+nx^{n+2}}{(x-1)^{2}}$$
Otras sumas analÃ­ticas incluyen

$$(\sum_{k=0}^{\infty} x^{k})^{p} = (1-x)^{-p}$$$$= \frac{1}{(p-1)!} \sum_{n=0}^{\infty}(n+1)_{p-1}x^{n}$$
para $|x|<1$, donde $(n)_p$ es un sÃ­mbolo de Pochhammer. La versiÃ³n finita tiene la elegante forma cerrada.

$$(\sum_{k=0}^{\infty} x^{k})^{p} = \frac{1}{(p-1)!} \sum_{k=0}^{np} \frac{(n-|n-k|+p-1)!}{(n-|n-k|)!}x^{k}$$
para $p=1$ y 2. Una suma adicional viene dada por

$$(\sum_{n=0}^{\infty} a_{n}x^{n})^{2} = \sum_{n=0}^{\infty} a_{n}^{2}x^{2n} + 2 \sum_{n=1}^{\infty}a_{i}a_{j}x^{n}$$
Una soluciÃ³n analÃ­tica para una suma de potencias de nÃºmeros enteros es

$$S_{p}(n) = \sum_{k=1}^{n}k^{p} = H_{n}^{-p}$$
Bernoulli usÃ³ la propiedad del triÃ¡ngulo numÃ©rico figurado que

$$\sum_{i=0}^{n}a_{ij} = \frac{(n+1)a_{nj}}{j+1}$$
junto con un form para a_(nj) que derivÃ³ inductivamente para calcular las sumas hasta n=10 (Boyer 1968, p. 85). Para p en $Z>0$, la suma estÃ¡ dada por

$$\sum_{k=1}^{n}k^{p} = \frac{(B + n + 1)^{(p+1)}-B^{(p+1)}}{p+1}$$
Donde la notaciÃ³n $B^k$ significa que la cantidad en cuestiÃ³n se eleva a la potencia apropiada k, y todos los tÃ©rminos de la forma $ B ^ m $ se reemplazan con los nÃºmeros de Bernoulli correspondientes $ B_m $. Escrito explÃ­citamente en tÃ©rminos de una suma de potencias,

$$\sum_{k=1}^{n} k^{p} = \sum_{k=1}^{p+1}b_{pk}n^{k}$$
where

$$b_{pk} = \frac{(-1)^{p-k+1}B_{p-k+1}p!}{k!(p-k+1)!}$$
TambiÃ©n es cierto que los coeficientes de los tÃ©rminos b_ (pk) suman 1,

$$\sum_{k=1}^{p+1}b_{pk}=1$$
que Bernoulli declarÃ³ sin prueba.
